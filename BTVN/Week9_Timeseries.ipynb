a) Hãy phân tích mô hình ARMA(1,1) với tham số $
\theta_1=0.21; \phi_1=0.09;$

```python
ar = np.array([0.09])
ma = np.array([0.21])
ARMA_vd1=ArmaProcess.from_coeffs(ar, ma)
```

```python
print('Tinh khả nghịch:', ARMA_vd1.isinvertible)
print('Tính dừng của chuỗi thời gian:', ARMA_vd1.isstationary)
print('Đa thức phần MA:', ARMA_vd1.mapoly)
print('Nghiệm của đa thức phần MA', ARMA_vd1.maroots)
print('Đa thức phần AR:', ARMA_vd1.arpoly)
print('Nghiệm của đa thức phần MA', ARMA_vd1.arroots)
print('Các giá trị acf:', np.round(ARMA_vd1.acf(10),4)) 
print('Các giá trị acvf:', np.round(ARMA_vd1.acovf(10),4)) 
print( ' Chuyển ARMA thành MA', ARMA_vd1.arma2ma(10))
print(' Chuyển ARMA thành AR', ARMA_vd1.arma2ar(10))
```

```python
ARMA_data = ARMA_vd1.generate_sample(nsample=10000)
```

```python
plt.figure(figsize=[15, 7.5]); # Set dimensions for figure
plt.plot(ARMA_data)
plt.title("Mô phỏng quá trình  ARMA_vd1 ")
plt.xlim([0, 200])
plt.show()
```

```python
plot_pacf(ARMA_data);
```

```python
plot_acf(ARMA_data);
```

b) Hãy phân tích mô hình ARMA(1,1) với tham số $
\theta_1=0.09; \theta_2=0.21;  \phi_1=0.21; \phi_2=0.09$

```python
ar = np.array([0.21, 0.09])
ma = np.array([0.09,0.21])
ARMA_vd1=ArmaProcess.from_coeffs(ar, ma)
```

```python

print('Tinh khả nghịch:', ARMA_vd1.isinvertible)
print('Tính dừng của chuỗi thời gian:', ARMA_vd1.isstationary)
print('Đa thức phần MA:', ARMA_vd1.mapoly)
print('Nghiệm của đa thức phần MA', ARMA_vd1.maroots)
print('Đa thức phần AR:', ARMA_vd1.arpoly)
print('Nghiệm của đa thức phần MA', ARMA_vd1.arroots)
print('Các giá trị acf:', np.round(ARMA_vd1.acf(10),4)) 
print('Các giá trị acvf:', np.round(ARMA_vd1.acovf(10),4)) 
print( ' Chuyển ARMA thành MA', ARMA_vd1.arma2ma(10))
print(' Chuyển ARMA thành AR', ARMA_vd1.arma2ar(10))
```

```python
ARMA_data = ARMA_vd1.generate_sample(nsample=10000)
```

```python
plt.figure(figsize=[15, 7.5]); # Set dimensions for figure
plt.plot(ARMA_data)
plt.title("Mô phỏng quá trình  ARMA_vd1 ")
plt.xlim([0, 200])
plt.show()
```
